```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay

from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.neighbors import KNeighborsClassifier

from tensorflow.keras.datasets import cifar10
```

```python
(X_train_full, y_train_full), (X_test_full, y_test_full) = cifar10.load_data()

y_train_full = y_train_full.flatten()
y_test_full = y_test_full.flatten()

print("Train shape:", X_train_full.shape)
print("Test shape:", X_test_full.shape)
```

```python
# reduce dataset size to make training faster

X_train_full = X_train_full[:3000]
y_train_full = y_train_full[:3000]

X_test_full = X_test_full[:500]
y_test_full = y_test_full[:500]

print("Reduced train:", X_train_full.shape)
print("Reduced test:", X_test_full.shape)
```

```python
X = X_train_full.reshape(X_train_full.shape[0], -1) / 255.0
X_test = X_test_full.reshape(X_test_full.shape[0], -1) / 255.0
```

```python
X_train, X_val, y_train, y_val = train_test_split(
    X, y_train_full,
    test_size=0.2,
    random_state=42,
    stratify=y_train_full
)

print("Train:", X_train.shape)
print("Validation:", X_val.shape)
print("Test:", X_test.shape)
```

```python
logreg = Pipeline([
    ("scaler", StandardScaler()),
    ("model", LogisticRegression(max_iter=300, n_jobs=-1))
])

logreg.fit(X_train, y_train)

pred_val = logreg.predict(X_val)

acc = accuracy_score(y_val, pred_val)

print("Logistic Regression validation accuracy:", acc)
```

```python
from sklearn.linear_model import SGDClassifier

svm = Pipeline([
    ("scaler", StandardScaler()),
    ("model", SGDClassifier(loss="hinge", max_iter=1000, random_state=42))
])

svm.fit(X_train, y_train)

pred_val = svm.predict(X_val)
acc = accuracy_score(y_val, pred_val)

print("SVM validation accuracy:", acc)
```

```python
knn = Pipeline([
    ("scaler", StandardScaler()),
    ("model", KNeighborsClassifier(n_neighbors=5))
])

knn.fit(X_train, y_train)

pred_val = knn.predict(X_val)

acc = accuracy_score(y_val, pred_val)

print("KNN validation accuracy:", acc)
```

```python
pred_test = knn.predict(X_test)

acc_test = accuracy_score(y_test_full, pred_test)

print("KNN test accuracy:", acc_test)
```

```python
pred_test = svm.predict(X_test)

acc_test = accuracy_score(y_test_full, pred_test)

print("SVM test accuracy:", acc_test)
```
